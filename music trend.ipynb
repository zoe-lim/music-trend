<h1> 3. Exploratory data analysis </h1>
<h2> 3.1 Preference of Music Genre over the years </h2>

In order to track the preference of music genres over the years, the code below chooses the top 10 most popular genre of each year from the year 2010 to 2019. 

```python
genres_year = data[['Genre', 'Top_Year']]

def genre_type(ds):
    years = range(2010, 2020)
    y_g = pd.DataFrame()
    
    for i in years:
        # select top 10 most popular genre of every year
        col = ds[ds.Top_Year == i]['Genre'].value_counts().head(10) 
        col.to_frame()
        col = col.rename(str(i))
        y_g = y_g.append(col)

    return y_g

genres_years = genre_type(genres_year)
genres_years
```

```python
# plot top 10 most popular genre every year against year graph
f1 = plt.figure()
genres_years.plot(kind='bar', ax=f1.gca(), stacked=True, figsize=(12,10), ylabel='Number of Songs in the Genre', xlabel='Years', title='Top 10 Most Popular Genres of the Year')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

<h3> 1. Dance Pop </h3>

As can be seen from the bar graph above, the most popular music genre of all time is dance pop. However, the graph also shows a decline in the number of tracks with the genre dance pop over the years. This shows that despite dance pop being the most popular genre, it is also slowly losing popularity. This could be due to other music genres gaining popularity, which will be discussed in the section below, or that there are fewer dance pop tracks being released. 

<h3> 2. Atl Hip Hop </h3>

Although not as popular as dance pop, atl hip hop was also relatively quite popular in the year 2010, however, in the year 2011, there was a drastic decrease in the number of tracks with the genre atl hip hop. However, in the year 2017, there was a gain in the number atl hip hop tracks, which shows that this particular genre may still continue to gain popularity in the future.

<h3> 3. Pop </h3>

As opposed to the two genres above, the pop genre is increasing over the years. The increase is gradual, showing that the genre is slowly gaining popularity over the years and may become the new trendy music genre in the future.

<h2> 3.2 Preference of Music Attributes over the years </h2>
<h3> 3.2.1 Correlations between Music Attributes and Popularity </h3>

Below is a correlation table created with a heatmap to investigate the possible correlations between music attributes and the popularity of the track.

```python
# create heatmap with seaborn
plt.figure(figsize=(16, 8))
sns.set(style="whitegrid")
corr = data.corr()
sns.heatmap(corr,annot=True, cmap="PuRd")
```

Looking at the correlation table, we can note a few observations on what attributes make a song more popular.
<ul>
    <li>
        Danceability seems to be slightly correlated to popularity with a value of 0.093.
    </li>
    <li>
        Acousticness has a slightly higher value of the correlation of 0.13 with popularity.
    </li>
    <li>
        Year released has a correlation value of 0.18 with popularity, meaning that the newer the song is, the higher the chance of it being popular.
    </li>
    <li>
        Energy has a negative correlation value of -0.23 with popularity. This means that the tracks are more popular when they have lesser energy.
    </li>
</ul>
However, it can also be seen that all the correlation values with popularity are rather low, and none of the values is highly correlated. This means that the popularity of a track is not dependent on the music attributes. 

<h3> 3.2.2 Trends of the Different Music Attributes </h3>
<h3> 1. Energy </h3>

```python
plt.figure(figsize=(15, 5))

# mean energy over year
x = data.groupby('Top_Year')["Energy"].mean()
ax = sns.lineplot(x= x.index, y=x, label="Energy")

# set axis label and titles 
ax.set_title('Mean of Energy of Music over the Years', fontsize = 15)
ax.set_ylabel('Energy', fontsize = 15)
ax.set_xlabel('Year', fontsize = 15)
```

As shown in graph above, the energy of music is decreasing over the years, showing that people are now more likely to prefer a song with lower energy as compared to the past.

<h3> 2. Valence </h3>

```python
plt.figure(figsize=(15, 5))

# mean energy over year
x = data.groupby('Top_Year')["Valence"].mean()
ax = sns.lineplot(x= x.index, y=x, label="Valence")

# set axis label and titles 
ax.set_title('Mean of Valence of Music over the Years', fontsize = 15)
ax.set_ylabel('Valence', fontsize = 15)
ax.set_xlabel('Year', fontsize = 15)
```

The valence value shows the mood of a song, the higher the value, the more positive mood the song is. Hence, the graph above shows that over the years people seem to prefer songs that are sadder as compared to the past. However, there is an increase in the value after the year 2015. This shows that the music preference of people seems to have changed once again, and start to prefer happier songs. 

<h3> 3. Acousticness </h3>

```python
plt.figure(figsize=(15, 5))
# mean acousticness over year
x = data.groupby('Top_Year')["Acousticness"].mean()
ax = sns.lineplot(x= x.index, y=x, label="Acousticness")

# set axis label and titles 
ax.set_title('Mean Acousticness of Music over the Years', fontsize = 15)
ax.set_ylabel('Acousticness', fontsize = 15)
ax.set_xlabel('Year', fontsize = 15)
```

The trend of acousticness is increasing over the years, which means a gain in popularity of songs which are more acoustic.

<h3> 4. Danceability </h3>

```python
plt.figure(figsize=(20, 5))

# mean loudness over year
x = data.groupby('Top_Year')["Danceability"].mean()
ax = sns.lineplot(x= x.index, y=x, label="Danceability")

# set axis label and titles 
ax.set_title('Mean Danceability of Music over the Years', fontsize = 15)
ax.set_ylabel('Danceability', fontsize = 15)
ax.set_xlabel('Year', fontsize = 15)
```

The value of danceability is increasing over the years, showing that people now prefer songs that are easy to dance to.

<h3> 5. BPM </h3>

```python
plt.figure(figsize=(15, 5))

# mean BPM over year
x = data.groupby('Top_Year')["BPM"].mean()
ax = sns.lineplot(x= x.index, y=x, label="BPM")

# set axis label and titles 
ax.set_title('Mean BPM of Music over the Years', fontsize = 15)
ax.set_ylabel('BPM', fontsize = 15)
ax.set_xlabel('Year', fontsize = 15)
```

There are ups and downs in the BPM value over the years, however, the general trend is a gradual increase. This shows that the BPM does not play a big part in deciding the popularity of a song. 

<h2> 3.3 Popular Themes of Music based on Title </h2>

Below shows some popular words used in the title of the songs in the Top Hits playlist over the years.

```python
text = " ".join(Title for Title in data.Title)
text = "".join(re.split("\(|\)|\[|\]", text)[::2])
text = text.replace('Edit ', '')
text = text.replace('Radio ', '')
text = text.replace('Remix', '')

word_cloud = WordCloud(collocations = False, background_color = 'white').generate(text)

plt.imshow(word_cloud, interpolation='bilinear')
plt.title('Map of the most common words in song titles')
plt.axis("off")
plt.show()
```

As can be seen above, the most popular word used is "Love". Hence, it can be seen that people relate most to songs that are talking about love and have a preference for these songs.
